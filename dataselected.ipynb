```python
df = pd.read_csv('seleted10.csv')
calm_data = df[df['emotion'] == '平静']
calm_data
```

```python
feature_columns = calm_data.select_dtypes(include='number').columns.tolist()
feature_columns
```

```python
 # 3. 分组与聚合：按 'subject' 分组，并计算均值和标准差
baseline_stats = calm_data.groupby('subject')[feature_columns].agg(['mean', 'std'])
baseline_stats 
```

```python
# 4. Z-score 归一化
# 创建一个副本用于存储归一化后的数据
df_normalized = df.copy()

# 获取要归一化的特征列
feature_columns_to_normalize = feature_columns # 这个变量在之前的单元格中已经定义

# 遍历每个被试
for subject in df['subject'].unique():
    # 检查该被试是否存在于基线统计数据中
    if subject in baseline_stats.index:
        # 获取当前被试的均值和标准差
        subject_means = baseline_stats.loc[subject].loc[(feature_columns_to_normalize, 'mean')]
        subject_stds = baseline_stats.loc[subject].loc[(feature_columns_to_normalize, 'std')]
        
        # 提取索引，以便正确赋值
        subject_means.index = subject_means.index.get_level_values(0)
        subject_stds.index = subject_stds.index.get_level_values(0)

        # print(subject_means, '\n',subject_means.index)

        # 定位当前被试在原始数据中的所有行
        subject_indices = df_normalized['subject'] == subject
        
        # 对每个特征列进行z-score归一化
        # z = (x - mean) / std
        df_normalized.loc[subject_indices, feature_columns_to_normalize] = \
            (df.loc[subject_indices, feature_columns_to_normalize] - subject_means) / subject_stds

print("Z-score 归一化完成。")
df_normalized.head()
```

```python
df_normalized.to_csv('seleted10_normalized.csv', index=False)
```


### 验证数据处理的正确性

为了确保Z-score归一化过程是正确且可逆的，我们将执行以下反向验证步骤：

1.  **选择一个样本**：我们选取一名被试在特定情绪下的几条数据作为样本。
2.  **获取归一化数据和基线统计**：从`df_normalized`中提取样本的归一化特征值，并从`baseline_stats`中获取该被试的均值（mean）和标准差（std）。
3.  **反向计算**：应用Z-score的逆公式 `原始值 = (Z-score * 标准差) + 均值`，来还原原始数据。
4.  **对比验证**：将还原后的数据与`seleted10.csv`中的原始数据进行比较。

如果还原后的数据与原始数据基本一致（考虑到浮点数计算可能存在的微小误差），则证明我们的数据清洗和归一化流程是正确的。


```python

# 验证归一化过程的正确性
df = pd.read_csv('seleted10.csv')
df_normalized = pd.read_csv('seleted10_normalized.csv')
# 1. 选择一个被试和情绪进行验证
subject_to_verify = '赵紫涵'
emotion_to_verify = '愉悦'
# 选择要验证的特征列
feature_cols_to_verify = ['RMSSD', 'pNN58', 'SDNN', 'SD1', 'SD2', 'SD1_SD2']

# 2. 从归一化数据中提取样本
normalized_sample = df_normalized[
    (df_normalized['subject'] == subject_to_verify) & 
    (df_normalized['emotion'] == emotion_to_verify)
][feature_cols_to_verify].head(1) # 取第一个样本点

# 3. 从原始数据中提取对应的样本
original_sample = df[
    (df['subject'] == subject_to_verify) & 
    (df['emotion'] == emotion_to_verify)
][feature_cols_to_verify].head(1)

# 4. 获取该被试的基线统计数据 (均值和标准差)
subject_means = baseline_stats.loc[subject_to_verify].loc[(feature_cols_to_verify, 'mean')]
subject_stds = baseline_stats.loc[subject_to_verify].loc[(feature_cols_to_verify, 'std')]

# 移除多重索引以便计算
subject_means.index = subject_means.index.get_level_values(0)
subject_stds.index = subject_stds.index.get_level_values(0)

# 5. 执行反向计算： original = (z_score * std) + mean
restored_data = (normalized_sample * subject_stds) + subject_means

# 6. 打印结果进行对比
print(f"正在验证被试 '{subject_to_verify}' 在情绪 '{emotion_to_verify}' 下的数据")
print("-" * 50)
print("原始数据:")
print(original_sample)
print("-" * 50)
print("反向计算还原的数据:")
# 重置索引以匹配格式
restored_data.index = original_sample.index
print(restored_data)
print("-" * 50)
print("结论：比较以上两组数据，如果数值非常接近（可能存在微小的浮点计算误差），则证明归一化过程正确无误。")
```
